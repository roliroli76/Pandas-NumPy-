```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as math
```

```python
%matplotlib inline
```

```python
days = [1, 2, 3, 4, 5]
amount = [1, 5, 20, 450, 700]
plt.plot(days, amount)
plt.xlabel('days')
plt.ylabel('amount')
plt.show()
```

```python
days = [1, 2, 3, 4, 5]
amount = [1, 5, 20, 450, 700]
plt.plot(days, amount, 'bs')
plt.xlabel('days')
plt.ylabel('amount')
plt.show()
```

```python
months = np.arange(1, 13, 1)
months

```

```python
income = 100 + 3 * months
income
```

```python
plt.plot(months, income, 'go')
```

```python
cost = 50 + 10 * months
```

```python
plt.plot(months, income, 'go', months, cost, 'r^')
```

```python
xArguments = np.arange(-5, 5, 0.1)
yValues = 2**xArguments
plt.plot(xArguments, yValues)
plt.xlabel('arguments')
plt.ylabel("values")
```
